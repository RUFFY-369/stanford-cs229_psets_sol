```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r'ds4_train.csv')
```

```python
x_train=data.iloc[:,0:4]
y_train=data.loc[:,['y']]

xv=x_train.values
colsx=x_train.shape[0]
zer=np.ones((colsx,1))
x_train_new=np.hstack((zer,xv))

```

```python
itera=0
col_x=(x_train_new.shape[1])
theta1=np.zeros((col_x,1))
lr=1e-7
m=x_train_new.shape[0]
g=lambda x:np.exp(x)
while itera<2500:
    theta=theta1

    grad=(1/m)*((x_train_new).T).dot(((y_train)-g((x_train_new.dot(theta)))))
    

    theta1=theta+(grad*lr)
    
    dell=np.linalg.norm((theta1-theta),ord=1)

    if dell< 1e-8:
        print(f'break up party at {itera}th iteration')
        break
        
    
    itera+=1

theta=theta1
print(theta)


y_pred=np.exp(((x_train_new)@theta))
print(y_pred)
```
